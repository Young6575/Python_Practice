## ì‹¤ìŠµ ë¬¸ì œ: í–‰ë ¬ ì „ì¹˜(Transpose) êµ¬í˜„í•˜ê¸°

**ì„¤ëª…**: í–‰ë ¬ ì „ì¹˜ëŠ” í–‰ë ¬ì˜ í–‰ê³¼ ì—´ì„ ì„œë¡œ ë§ë°”ê¾¸ëŠ” ì—°ì‚°ì…ë‹ˆë‹¤. ì¦‰, ì›ë³¸ í–‰ë ¬ì˜ ì²« ë²ˆì§¸ í–‰ì€ ì „ì¹˜ í–‰ë ¬ì˜ ì²« ë²ˆì§¸ ì—´ì´ ë˜ê³ , ë‘ ë²ˆì§¸ í–‰ì€ ë‘ ë²ˆì§¸ ì—´ì´ ë˜ëŠ” ì‹ì…ë‹ˆë‹¤. ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ” ìˆœìˆ˜ íŒŒì´ì¬ì„ ì‚¬ìš©í•˜ì—¬ ì£¼ì–´ì§„ í–‰ë ¬ì„ ì „ì¹˜í•˜ëŠ” í•¨ìˆ˜ë¥¼ êµ¬í˜„í•©ë‹ˆë‹¤. ğŸ“

$$(A^T)_{ij} = A_{ji}$$

**ìš”êµ¬ì‚¬í•­**:

  - `transpose` í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì—¬ ì¸ìë¡œ ë°›ì€ `matrix`ì˜ í–‰ê³¼ ì—´ì„ ë’¤ë°”ê¾¼ ìƒˆë¡œìš´ í–‰ë ¬ì„ ë°˜í™˜í•˜ì„¸ìš”.
  - ì›ë³¸ í–‰ë ¬ì˜ `i`ë²ˆì§¸ í–‰, `j`ë²ˆì§¸ ì—´ì˜ ìš”ì†ŒëŠ” ìƒˆ í–‰ë ¬ì˜ `j`ë²ˆì§¸ í–‰, `i`ë²ˆì§¸ ì—´ì— ìœ„ì¹˜í•´ì•¼ í•©ë‹ˆë‹¤.
  - í•¨ìˆ˜ë¥¼ í˜¸ì¶œí•˜ì—¬ ì›ë³¸ í–‰ë ¬ê³¼ ì „ì¹˜ëœ í–‰ë ¬ì„ ëª¨ë‘ ì¶œë ¥í•˜ì—¬ ê²°ê³¼ê°€ ì˜¬ë°”ë¥¸ì§€ í™•ì¸í•˜ì„¸ìš”.


```python
def transpose(matrix: list[list[int]]) -> list[list[int]]:
    """
    ì£¼ì–´ì§„ í–‰ë ¬ì˜ í–‰ê³¼ ì—´ì„ ë’¤ë°”ê¾¼ ì „ì¹˜ í–‰ë ¬ì„ ë°˜í™˜í•©ë‹ˆë‹¤.

    Args:
        matrix (list[list[int]]): ì „ì¹˜í•  2ì°¨ì› ë¦¬ìŠ¤íŠ¸ í˜•íƒœì˜ í–‰ë ¬.

    Returns:
        list[list[int]]: ì „ì¹˜ëœ ìƒˆë¡œìš´ í–‰ë ¬.
    """
    result = [[0 for _ in range(len(matrix))] for _ in range(len(matrix[0]))]

    for i in range(len(matrix[0])):
        for j in range(len(matrix)):
            result[i][j] = matrix[j][i]
    return result
        
        # return list(zip(*A))

```
