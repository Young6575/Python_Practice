```python
cars = ['bmw' , 'genesis' , 'sorento' , 'k5' , 'benz']
cars.reverse()
print(cars) 
cars.sort(reverse=True) # list 자체를 정렬 / sort가 매서드 임으로 정렬 권한이 있음.
# list 자체는 정렬하지 않고 정렬된 결과만 출력
#print(sorted(cars)) # 메소드가 아니고 라이브러리 함수 / 
print(cars)
cars.sort()
print(cars) 
```

```python
print(cars[-1]) # -1은 마지막 요소 > 아주 편리한 데 복잡한 코딩
print(cars[1:-2]) # -2 마지막에서 두번째 요소
print(cars[1:3])
```

```python
print(cars)
#loop block의 시작과 끝(들여쓰기가 끝나면 loop 끝)
for c in cars: # 확장형 for 문 => 자바 유사
    print(f"차 이름은 = {c}")
print("성공적으로 종료")
cars.reverse()
print(cars)
```
